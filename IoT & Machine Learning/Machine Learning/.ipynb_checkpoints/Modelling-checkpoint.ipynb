```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
alpha = pd.read_csv("alpha.csv")
beta = pd.read_csv("beta.csv")
charlie = pd.read_csv("charlie.csv")

data = alpha.append(beta).append(charlie).reset_index(drop=True)
```

```python
data.head()
```

```python
data.corr()['CO2'].drop(["CO2","CO"]).sort_values().plot(kind='bar', figsize=(7, 6))
```

```python
target = data.CO2
data = data.drop(["CO2", "CO"], axis = 1)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
from sklearn.model_selection import train_test_split

# Split the DataFrame into train and test sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# Print the shapes of the train and test sets
print("Train set shape:", train_data.shape)
print("Test set shape:", test_data.shape)
```

```python
sns.set()
y=target
y_transformed = pd.Series(np.log1p(y))
y_sqrt = y**0.5

fig, ax = plt.subplots(1, 2)
plt.title("No transformation")
sns.distplot(y, ax=ax[0])
plt.title("Target after Log transformation")
# ax[0].axvline(y_transformed)
sns.distplot(y_transformed, ax=ax[1])
#plt.title("Target after SQRT transformation")
#sns.distplot(y_sqrt, ax=ax[1])
plt.show()
```

```python
X = train_data
y = target[:len(train_data)]
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import KFold
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
```

```python
fold_pred=[]
oof_pred = []

"""params = {#'colsample_bytree': 0.384606276881856,
 'learning_rate': 0.163251453473545997,
 'max_depth': 14,
 'subsample': 0.6530887571489526}
"""
params = {}#'learning_rate': 0.024579949101979186, 'subsample': 0.7970980112408316, 'colsample_bytree': 0.30922095499982644, 'max_depth': 16, 'n_estimators': 1117} #{'learning_rate': 0.027692082452321506, 'subsample': 0.48402420041502403, 'colsample_bytree': 0.13077221300845826, 'max_depth': 10, 'n_estimators': 1090}

#{'learning_rate': 0.024579949101979186, 'subsample': 0.7970980112408316, 'colsample_bytree': 0.30922095499982644, 'max_depth': 16, 'n_estimators': 1117}

#top_x_features = 90

fold= KFold(n_splits=15, shuffle=True, random_state = 2023)#15#5#10
i=1
for train_index, test_index in fold.split(X,y):     
  
    #X_train, X_test = X[lgbm_feature_importance[:top_x_features]].iloc[train_index], X[lgbm_feature_importance[:top_x_features]].iloc[test_index]
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    #y_train, y_test = np.log1p(y.iloc[train_index]), y.iloc[test_index]
    #y_train, y_test = np.sqrt(y.iloc[train_index]), y.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model = CatBoostRegressor(**params)
    model.fit(X_train,y_train,eval_set=[(X_train,y_train),(X_test, y_test)], early_stopping_rounds=200, verbose = False)#erly100

    preds= model.predict(X_test)
    print("err: ",(mean_squared_error(y_test,preds))**0.5)  #Reverse transformation
    oof_pred.append(mean_squared_error(y_test,preds)**0.5)
    #print("err: ",(mean_squared_error(y_test,preds**2))**0.5)  #Reverse transformation
    #oof_pred.append(mean_squared_error(y_test,preds**2)**0.5)
    #print("err: ",(mean_squared_error(y_test,np.expm1(preds)))**0.5)  #Reverse transformation
    #oof_pred.append(mean_squared_error(y_test,np.expm1(preds))**0.5)
    #p2 = model.predict(df_test[lgbm_feature_importance[:top_x_features]])
    p2 = model.predict(test_data[X.columns])
    #fold_pred.append(np.expm1(p2))
    fold_pred.append(p2)
    #fold_pred.append(p2**2)
    

print(np.mean(oof_pred)) #9.625137172354112 #9.61 9.53351001575291  9.515475216155535 9.4720906063346
```

```python
# # 0.7831404345536787
feature_importance_df = pd.DataFrame(model.feature_importances_, columns=['importance'])
feature_importance_df['feature'] = X.columns#lgbm_feature_importance[:top_x_features]# 
sns.set(font_scale = 2)

plt.figure(figsize=(20, 12));
sns.barplot(x="importance", y="feature", data=feature_importance_df.sort_values(by = ['importance'], ascending = False).head(60))
plt.title('CatboostClassifier features importance (top 50):')
```

```python
import matplotlib.pyplot as plt

y_actual = target[len(train_data):]
y_pred = np.mean(fold_pred, axis = 0)

# Create a scatter plot
plt.scatter(y_actual, y_pred)

# Add labels and title
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Scatter Plot of Predicted vs. Actual Values')

# Add a 45-degree reference line
plt.plot([min(y_actual), max(y_actual)], [min(y_actual), max(y_actual)], color='red')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve

# Assuming you have a regression model called 'model' and your features 'X' and target 'y'
# Adjust the names of the variables accordingly to match your data

# Set up the learning curve parameters
train_sizes, train_scores, test_scores = learning_curve(model, data, target, train_sizes=np.linspace(0.1, 1.0, 10), cv=5)

# Calculate the mean and standard deviation of the training and test scores
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

# Plot the learning curve
plt.figure(figsize=(8, 6))
plt.plot(train_sizes, train_mean, label='Training Score', color='blue')
plt.plot(train_sizes, test_mean, label='Cross-Validation Score', color='red')

# Add error bars
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color='lightblue')
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, color='pink')

# Customize the plot
plt.title('Learning Curve')
plt.xlabel('Training Set Size')
plt.ylabel('Score')
plt.legend(loc='best')
plt.grid(True)

# Show the plot
plt.show()
```
